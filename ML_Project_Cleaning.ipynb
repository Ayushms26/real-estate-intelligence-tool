# Now convert all price into 'Crore' scale

```python
def price(x):
    if isinstance(x, float) or pd.isna(x):
        return x
    parts = x.split(' ')
    if len(parts) < 2:
        return np.nan  # or handle it differently
    if parts[1] == 'Lac':
        return round(float(parts[0]) / 100, 2)
    else:
        return round(float(parts[0]), 2)

df['price'] = df['price'].apply(price)
```

```python

df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
